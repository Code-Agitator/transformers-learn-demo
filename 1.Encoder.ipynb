# 编码器
- 把文本编码成数字的工具

```python
sentence = 'hello everyone , today is a good day .'
```

- 字典，也叫做词表 vocabulary

```python
vocab = {
    '<SOS>': 0,  # start of sentence
    '<EOS>': 1,  # end of sentence
    'hello': 2,
    'everyone': 3,
    'today': 4,
    'is': 5,
    'a': 6,
    'good': 7,
    'day': 8,
    ',': 9,
    '.': 10
}
```

```python
sent = f'<SOS> {sentence} <EOS>'
sent
```

```python
# 英文分词，比较简单，直接按照空格分区
# 中文分词可以使用一些工具 例如jieba分词
words = sent.split()
words
```

```python
[vocab[i] for i in words]
```

# 1 使用编码工具

```python
# 模型和他的编码器是成对使用的， 使用什么模型 就用它提供的编码器
# 编码器的名字一般和模型的名字一样
# bert-base-chinese
from transformers import BertTokenizer
```

```python
tokenizer = BertTokenizer.from_pretrained(
    pretrained_model_name_or_path='bert-base-chinese',
    cache_dir='.cache',
    force_download=False
)
```

```python
sents = [
    '你站在桥上看风景',
    '看风景的人在楼上看你',
    '明月装饰了你的窗子',
    '你装饰了别人的梦'
]
```

```python
# 基本的编码函数
out = tokenizer.encode(
    text=sents[0],  # 主要句子
    text_pair=sents[1],  # 句子对 可以为空
    truncation=True,  # 截断，句子长度超出限制时，截断
    padding='max_length',  #  填充，句子长度不足限制时，填充
    add_special_tokens=True,  # 添加特殊标记
    return_tensors=None,  # 返回张量，可选项：None, 'pt', 'tf'
    max_length=25
)
out
```

```python
tokenizer.decode(out)
```

# 2 进阶版的编码工具

```python
from transformers.tokenization_utils_base import TruncationStrategy

out = tokenizer.tokenize(
    text=sents[0],
    pair=sents[1],
    truncation_strategy=TruncationStrategy.ONLY_FIRST,
    max_length=25,
    add_special_tokens=True,
    return_tensors=None,
    return_token_type_ids=True,
    return_attention_mask=True,
    return_special_tokens_mask=True,
    return_length=True,
)
out
```
